```python
table = pd.read_csv("VSRR_Provisional_Drug_Overdose_Death_Counts.csv", thousands=',')
table= table.rename(columns={'Data Value' : 'Data_value',
              'Percent Complete': "Percent_complete",
              'Percent Pending Investigation': "Percent_pending_investigation",
              'State Name': "State_name",
              'Footnote Symbol': "Footnote_symbol",
              'Predicted Value': "Predicted_value"})
table
```

```python
print(table.dtypes)
```

```python
table = table.convert_dtypes()
#print(table.dtypes)
```

```python
np.random.seed(0)
# Shuffle the dataset
table_shuffled = table.sample(frac=1).reset_index(drop=True)

# Define sizes
debug_size = 10000
working_size = 60000

# Split the dataset
debug_dataset = table_shuffled.iloc[:debug_size]
working_dataset = table.iloc[:working_size]

# Save the datasets to new CSV files
debug_dataset.to_csv("debug_dataset.csv", index=False)
working_dataset.to_csv("working_dataset.csv", index=False)
```

```python
def create_readme(team_members, problem_description, license_text, Label, features_description, data_sources,References):
    with open('README.md', 'w') as f:
        # Title
        f.write("# Drug Overdose Deaths\n\n")

        # Team members
        f.write("## Team Members\n")
        for member in team_members:
            f.write(f"- {member}\n")
        f.write("\n")

        # Problem description
        f.write("## Problem Description\n")
        f.write(problem_description + "\n\n")

        # Label
        f.write("## Label\n")
        f.write(Label + "\n\n")

        # Features description
        f.write("## Features Description\n")
        f.write(features_description + "\n\n")

        # Data sources
        f.write("## Data Sources\n")
        f.write(data_sources + "\n\n")

        # Referances
        f.write("## References\n")
        f.write(References + "\n\n")


        # License
        f.write("## License\n")
        f.write(license_text + "\n")



# Example usage
team_members = ['Hélène Rondey - hrondey2@illinois.edu', 'Yang-Hsuan Huang -  yhh4@illinois.edu', 'Catherine Li - sl182@illinois.edu', 'Akshat Bhardwaj - akshatb4@illinois.edu']
problem_description = "This project aims to solve the issue of cause of the drug overdose, and make a prediction may help regulate the drug industry by law and policy that decrease the potential for drug overdoses on a state-by-state basis in the United States, and inspire non-profit organizations deal with this problem in ensuring their resources are properly addressed to areas with the most horrible overdose situations."
license_text = "Copyright © 2023 University of Illinois."
Label = """
- **Predicted_value**: Predicted OD count Per state (OD Count are often incomplete and causes of death may be pending investigation)
"""
features_description = """
1. **Month and Year**: Time frame of the data (e.g., April 2015).
2. **Indicator (Drug Type)**: Type of drug involved in the observation.
3. **State**: Geographical location of the data collection.
4. **Percent Pending Investigation**: Percentage of cases still under investigation.
"""
data_sources = """
- **Climate Data Online (CDO)**: [Climate Data Online (CDO)](https://www.ncdc.noaa.gov/cdo-web/datasets)
"""
References = """
- **SHAP**: [Github](https://github.com/shap/shap)
- **SHAP**: [Article](https://papers.nips.cc/paper_files/paper/2017/file/8a20a8621978632d76c43dfd28b67767-Paper.pdf)

"""

create_readme(team_members, problem_description, license_text,Label, features_description, data_sources,References)
```
